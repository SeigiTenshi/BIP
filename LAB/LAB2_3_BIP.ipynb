```python
import skimage.io as io 
import matplotlib.pyplot as plt
import skimage as sk
import skimage.color as clr
import matplotlib as mat
import numpy as np
import scipy as sc
import glob


roof = io.imread('roof.jpg')
fig, ax = plt.subplots(2,2, figsize = (20,20))
ax[0,0].imshow(roof,cmap='gray')
ax[0,1].imshow(roof[::3,::3],cmap='gray')
roofG = sk.filters.gaussian(roof)
ax[1,0].imshow(roofG,cmap='gray')
ax[1,1].imshow(roofG[::3,::3],cmap='gray')
```

```python
DFT = np.fft.fft2(roof)
DFTlow = np.fft.fftshift(DFT)
Mask = np.zeros(roof.shape)
m,n = roof.shape
m = m//2
k=90
Mask[m-k:m+k,m-k:m+k] = 1
roofDFT =  Mask * DFTlow 
roofIDFT = np.fft.ifft2(roofDFT)
abs = np.abs(roofIDFT)
roofMoirée = abs[::3,::3]

plt.imshow(roofMoirée,cmap='gray')
```
