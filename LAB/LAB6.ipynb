```python
import skimage.io as io 
import matplotlib.pyplot as plt
import skimage as sk
import skimage.color as clr
import matplotlib as mat
import numpy as np
import scipy as sc
import glob
```

```python
squirrel = io.imread('squirrel.png')
plt.imshow(squirrel,cmap='gray')
```

```python
SNR = 60
ps = np.linalg.norm(squirrel)**2/(squirrel.shape[0]*squirrel.shape[1])
var = pow(10,(-SNR/10)) * ps
noise_squirrel = sk.util.random_noise(squirrel, clip='False',var = var )
plt.imshow(noise_squirrel,cmap='gray')
```

```python
w = 9
h = np.ones((w,w)) / (w*w)
mean_squirell = sc.ndimage.convolve(noise_squirrel,h)
plt.imshow(mean_squirell,cmap='gray')
```

```python
tv_squirrel = sk.restoration.denoise_tv_chambolle(noise_squirrel)
plt.imshow(tv_squirrel,cmap='gray')
```
