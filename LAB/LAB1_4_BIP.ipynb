```python
import skimage.io as io 
import matplotlib.pyplot as plt
import skimage as sk
import skimage.color as clr
import matplotlib as mat
import numpy as np

image = np.zeros((40,80,3)) #le 3 correspond au b du cours
plt.imshow(image)

image[0:20, 0:20, 0] = 255  #0=R, 1=G, 2=B
image[0:20, 0:20, 1] = 255

image[20:40, 0:20, 1] = 255

image[0:20, 20:40, 0] = 255
image[0:20, 20:40, 1] = 255
image[0:20, 20:40, 2] = 255

image[0:20, 40:60, 0] = 255
image[0:20, 40:60, 2] = 255

image[0:20, 60:80, 0] = 255

image[20:40, 20:40,1] = 255
image[20:40, 20:40,2] = 255

image[20:40, 40:60,2] = 255

plt.imshow(image)
```

```python
santamonica = io.imread('santamonica.jpg')
Vsantamonica = np.ravel(santamonica)
#plt.imshow(santamonica)
plt.hist(Vsantamonica,bins=100)
plt.show()
```

```python
haze = io.imread('haze.png')
ig, ax = plt.subplots(1,2, figsize = (10,10))
Vhaze = np.ravel(haze)
ax[0].imshow(haze)
ax[1].hist(Vhaze,bins=100)
plt.show() # ressemble à un dirac
```

```python
Mhaze = haze * 5
ig, ax = plt.subplots(1,2, figsize = (10,10))
VMhaze = np.ravel(Mhaze)
ax[0].imshow(Mhaze)
ax[1].hist(VMhaze,bins=100)
plt.show() # ressemble à un dirac
```

```python
Ehaze = sk.exposure.equalize_hist(haze)
ig, ax = plt.subplots(1,2, figsize = (10,10))
VEhaze = np.ravel(Ehaze)
ax[0].imshow(Ehaze)
ax[1].hist(VEhaze,bins=100)
plt.show()
```
