```python
import skimage.io as io 
import matplotlib.pyplot as plt
import skimage as sk
import skimage.color as clr
import matplotlib as mat
import numpy as np
import scipy as sc
import glob

groundtruth = io.imread('groundtruth.png')
observation = io.imread('observation.png')

fig, ax = plt.subplots(1,4, figsize = (40,40))
ax[0].imshow(groundtruth,cmap='gray')
ax[0].set_title('Groundtruth', fontstyle='italic',fontsize  = 50)
ax[1].imshow(observation,cmap='gray')
ax[1].set_title('Observation', fontstyle='italic',fontsize  = 50)
ax[2].hist(observation)
ax[2].set_title('Histograme', fontstyle='italic',fontsize  = 50) #moy = 75
T = 101
binary = observation > T
ax[3].imshow(binary,cmap='gray')
ax[3].set_title('Binary', fontstyle='italic',fontsize  = 50)

dice = sc.spatial.distance.dice(groundtruth.ravel()/255,binary.ravel())
print('for first method dice is = ', 1-dice)
plt.show()
```

```python
otsu = sk.filters.threshold_otsu(observation)
binary_otsu = observation > otsu
fig, ax = plt.subplots(1,2, figsize = (20,20))
im1 = ax[0].imshow(groundtruth,cmap='gray')
ax[0].set_title('Groundtruth', fontstyle='italic',fontsize  = 50)
im2 = ax[1].imshow(binary_otsu,cmap='gray')
ax[1].set_title('otsu', fontstyle='italic',fontsize  = 50)
fig.colorbar(im1,location='bottom')
fig.colorbar(im2,location='bottom')
dice = sc.spatial.distance.dice(groundtruth.ravel()/255,binary_otsu.ravel())
print('for first method dice otsu is = ', 1-dice)
```

```python
local = sk.filters.threshold_local(observation, block_size = 21 , offset=10)
binary_local = observation > local
plt.imshow(binary_local,cmap='gray')
plt.show()
```
