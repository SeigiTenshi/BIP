```python
import skimage.io as io 
import matplotlib.pyplot as plt
import skimage.color as clr
import matplotlib as mat
```

```python
f = io.imread("4.2.03.tiff")
plt.imshow(f)
```

L'image est en 512 par 512 ou utilisation de f.shape

```python
fg = clr.rgb2gray(f)
plt.imshow(fg,cmap=plt.cm.Greys) #ici il faut choisir la colormap nous même pour affichez l'image en noir et blanc
print('the pixel 0,0 have an intensity of')
print(fg[0,0])
print(fg[1,0:5])
print(fg[2,:])
```

```python
row = fg[200,:]
mat.pyplot.plot(row) #on peux ici reconnaitre le nez du singe
```
