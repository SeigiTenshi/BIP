```python
import skimage.io as io 
import matplotlib.pyplot as plt
import skimage as sk
import skimage.color as clr
import matplotlib as mat
import numpy as np
import scipy as sc

smiley = io.imread('smiley.png')
floatSmiley = sk.img_as_float(smiley)
smileyG = sk.filters.gaussian(floatSmiley)
h = np.array((1,-1),ndmin=2)
fig, ax = plt.subplots(1,5, figsize = (40,40))

ax[0].imshow(floatSmiley)
smileyH= sc.ndimage.convolve(floatSmiley,h)
ax[1].imshow(smileyG,cmap='gray')
ax[2].imshow(smileyH,cmap='gray')

kernel1 = np.ones((1,30))
kernel1 = kernel1/30
smileyKernel = sc.ndimage.convolve(floatSmiley,kernel1)
ax[3].imshow(smileyKernel,cmap='gray')
a = np.zeros((255,255))
a[128,128] = 1
psf = sc.ndimage.convolve(a,h)
ax[4].imshow(psf,cmap='gray')
```
