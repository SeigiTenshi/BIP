```python
import skimage.io as io 
import matplotlib.pyplot as plt
import skimage as sk
import skimage.color as clr
import matplotlib as mat
import numpy as np
import scipy as sc
import glob

images = glob.glob('s/*.png')
print(len(images))

fig, ax = plt.subplots(len(images),3, figsize = (40,40))

for i in range(len(images)):
    si = io.imread(images[i])
    DFT = np.fft.fft2(si)
    DFTlow = np.fft.fftshift(DFT)
    abs = np.absolute(DFTlow)
    ang = np.angle(DFTlow)
    ax[i,1].imshow(abs,cmap='gray')
    ax[i,2].imshow(ang,cmap='gray')
    ax[i,0].imshow(si,cmap='gray')
```
