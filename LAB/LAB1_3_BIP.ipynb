```python
import glob
import matplotlib.pyplot as plt
import matplotlib as mt
import skimage.io as io 

images = glob.glob('flowers/*.JPG')

print("nombre d'images",len(images))

fig, ax = plt.subplots(1, len(images), figsize = (10,10)) #aller lire la doc (complexe as fuck)

for i in range(len(images)):
    ax[i].imshow(io.imread(images[i]))

#a réutiliser souvent très utile
```
