```python
import skimage.io as io 
import matplotlib.pyplot as plt
import skimage as sk
import skimage.color as clr
import matplotlib as mat
import numpy as np
import scipy as sc
import glob
```

```python
moliets = io.imread('moliets.png')
mask = 240
fig, ax = plt.subplots(1,4, figsize = (40,40))
test = np.ravel(moliets)
im1 = ax[0].imshow(moliets)
bleu = moliets[:,:,2]
rbleu = np.ravel(bleu)
fig.colorbar(im1,location='bottom')
im2 = ax[1].imshow(bleu,cmap='gray')
ax[2].hist(rbleu,bins=256)
finall = bleu.copy()
finall[bleu<mask] = 0
ax[3].imshow(finall,cmap='gray')
plt.show()
```

```python
a2D = sk.morphology.disk(10)

closing = sk.morphology.binary_closing(finall,a2D)
dilatation = sk.morphology.binary_dilation(finall,a2D)
erosion = sk.morphology.binary_erosion(finall,a2D)
opening = sk.morphology.binary_opening(finall,a2D)
fig, ax = plt.subplots(1,4, figsize = (40,40))
ax[0].imshow(closing,cmap='gray')
ax[1].imshow(dilatation,cmap='gray')
ax[2].imshow(erosion,cmap='gray')
ax[3].imshow(opening,cmap='gray')
plt.show()
```

```python
label = sk.morphology.label(closing)
mesure = sk.measure.regionprops(label)
```

```python
fig, ax = plt.subplots(1,2, figsize = (40,40))
ax[0].imshow(label)
plt.show()
print(len(mesure))
```

```python
closing20 = sk.morphology.remove_small_objects(closing,26)
label = sk.morphology.label(closing20)
mesure = sk.measure.regionprops(label)
plt.imshow(label)
print(len(mesure))
```
